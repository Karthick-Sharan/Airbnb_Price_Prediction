```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from warnings import filterwarnings
filterwarnings("ignore")
sns.set_theme()
print("Imported")
```

```python
df = pd.read_csv('clean.csv')
df.columns
```

```python
df['Unnamed: 0']
```

```python
df.drop(columns=['Unnamed: 0','State'], inplace=True)
```

```python
df.info()
```

```python
df['Bathrooms'] = df['Bathrooms'].astype('object')
df['Bedrooms'] = df['Bedrooms'].astype('object')
df['Beds'] = df['Beds'].astype('object')
```

```python
x = df.drop('Price',1)
y = df['Price']
```

```python
x_num = x.select_dtypes(np.number)
x_cat = x.select_dtypes('object')
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

x_num = pd.DataFrame(sc.fit_transform(x_num), columns=x_num.columns, index=x_num.index)
x_num
```

```python
x_cat = pd.get_dummies(x_cat, drop_first=True)
x_cat
```

```python
x = pd.concat([x_num, x_cat], axis=1)
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size=0.2, random_state=10)
```

```python
print(xtrain.shape)
print(xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr_model = lr.fit(xtrain,ytrain)
```

```python
ypred = lr_model.predict(xtest)
```

```python
result = pd.DataFrame({'ytest':ytest, 'ypred':ypred})
result
```

```python
from sklearn import metrics
print('RMSE:', metrics.mean_squared_error(ytest, ypred, squared=False))
print('R2:', metrics.r2_score(ytest,ypred))
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gb_reg = GradientBoostingRegressor()
gb_reg_model = gb_reg.fit(xtrain,ytrain)
```

```python
ypred_gb = gb_reg_model.predict(xtest)
print('RMSE:', metrics.mean_squared_error(ytest, ypred_gb, squared=False))
print('R2:', metrics.r2_score(ytest,ypred_gb))
```

```python
result_gb = pd.DataFrame({'ytest':ytest, 'ypred':ypred_gb, 'err': abs(ytest-ypred_gb)})
result_gb
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(n_estimators=50, max_depth=10)
rfr_model = rfr.fit(xtrain,ytrain)
```

```python
ypred_rf = rfr_model.predict(xtest)
print('RMSE:', metrics.mean_squared_error(ytest, ypred_rf, squared=False))
print('R2:', metrics.r2_score(ytest,ypred_rf))
```
